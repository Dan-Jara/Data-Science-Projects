```python
canvas_url = 'https://canvas.okstate.edu/courses/135328/users'
driver.get(canvas_url)
WebDriverWait(driver, timeout=10).until(lambda d: d.find_element('css selector','tr.rosterUser.al-hover-container.StudentEnrollment td a.roster_user_name.student_context_card_trigger'))
```

```python
names_elem = driver.find_elements('css selector','tr.rosterUser.al-hover-container.StudentEnrollment td a.roster_user_name.student_context_card_trigger')
name_list = list()
for names in names_elem:
    name_list.append(names.text)
print(name_list)
```

```python
canvas_url = 'https://canvas.okstate.edu/courses/135328/users'
driver.get(canvas_url)
WebDriverWait(driver, timeout=10).until(lambda d: d.find_element('css selector','div.section'))
```

```python
section_elem = driver.find_elements('css selector','div.section')
section_list = list()
for sections in section_elem:
    section_list.append(sections.text)
print(section_list)
```

```python
canvas_url = 'https://canvas.okstate.edu/courses/135328/users'
driver.get(canvas_url)

WebDriverWait(driver, timeout=10).until(lambda d: d.find_element('xpath','/html/body/div[3]/div[2]/div[2]/div[3]/div[1]/div/div[2]/div/div[1]/div/div[2]/table/tbody/tr/td[4]/div'))
```

```python
section_elem = driver.find_elements('xpath','/html/body/div[3]/div[2]/div[2]/div[3]/div[1]/div/div[2]/div/div[1]/div/div[2]/table/tbody/tr/td[4]/div')
roles_list = list()
for roles in section_elem:
    roles_list.append(roles.text)
print(roles_list)
```

```python
student_name = pd.DataFrame(name_list)
student_name.columns =['student_name']
student_name
```

```python
class_section = pd.DataFrame(section_list)
class_section.columns =['class_section']
class_section
```

```python
class_role = pd.DataFrame(roles_list)
class_role.columns =['class_role']
class_role
```

```python
dfs = pd.concat([student_name,class_section,class_role], axis=1)
dfs
```
