```python
import pandas as pd

from lxml import html
from lxml import etree

import urllib3
```

```python
imdburl = "https://www.imdb.com/title/tt1950186/?ref_=nv_sr_srsg_0"

httpmng = urllib3.PoolManager()
resp = httpmng.request('GET', imdburl)
```

```python
tagtree = html.fromstring(resp.data)
```

```python
# Title, genre (all individually, and then as a group and print via a loop), and overall rating of the movie
xp1 = '//div[@class="sc-80d4314-1 fbQftq"]/h1'

imdb_title = tagtree.xpath(xp1)
imdb_title[0].text_content()
```

```python
xp2 = '//div[@class="sc-7ab21ed2-2 kYEdvH"]'

imdb_rating = tagtree.xpath(xp2)
imdb_rating[0].text_content()
```

```python
imdburl3 = "https://en.wikipedia.org/wiki/Statistics_of_the_COVID-19_pandemic_in_the_United_States"

httpmng = urllib3.PoolManager()
resp = httpmng.request('GET', imdburl3)
tagtree3 = html.fromstring(resp.data)
```

```python
c1 = []
for i in range(3,59):
    col1 = "/html/body/div[3]/div[3]/div[5]/div[1]/div[7]/div/div/div[4]/table/tbody/tr["+str(i)+"]/th"
    wikipedia_table = tagtree3.xpath(col1)
    c1.append(wikipedia_table[0].text_content())

c1_new = []
for i in c1:
    c1_new.append(i.strip())

c1_new
```
