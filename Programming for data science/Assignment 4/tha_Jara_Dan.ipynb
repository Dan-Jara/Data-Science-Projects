```python
import pandas as pd
import regex
```

```python
hospital = pd.read_csv("data/CaliforniaHospitalData.csv")
personnel = pd.read_table("data/CaliforniaHospitalData_Personnel.txt")
```

```python
df = pd.concat([hospital,personnel], axis = 1)
df
```

```python
representatives_list = df["PositionTitle"].str.extract("(Regional Representative)")
representatives_list = representatives_list.dropna()
representatives_list = df.iloc[representatives_list.index]
representatives_list
```

```python
emily = df[df["FirstName"].str.contains("(Emily)")|df["LastName"].str.contains("(^A\w*|\w*n$)")]
emily = emily.dropna()
emily
```

```python
criteria = df[df["TypeControl"].str.contains("(Non Profit)|(Investor)") & df["Zip"].str.contains("(2$|4$|5$|6$)")]
criteria[['Zip','TypeControl', 'LastName', 'FirstName', 'PositionTitle']]
```

```python
criteria = df[df["TypeControl"].str.contains("(Non Profit)|(Investor)") & 
              df["Zip"].str.contains("(2$|4$|5$|6$)") & ~ df["PositionTitle"].str.contains("(Safety Inspection Member)")]
criteria[['Zip','TypeControl', 'LastName', 'FirstName', 'PositionTitle']]
```
