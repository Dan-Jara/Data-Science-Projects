#### Instrucción 3: 
<p style='text-align: justify;'>En la siguiente sección de nuestra libreta  analizaremos a la variable "<b>Body mass index</b>" (bmi). Crea una variable <b>x</b> donde grabes la columna <b>bmi</b> y una variable <b>y</b> donde almacenemos la información de la columna <b>target</b>. <b>Nota:</b> es muy recomendable a cada una de las variables les des la forma de una columna de datos con ayuda de la instrucción <b>values.reshape(-1, 1)</b>.</p>

```python
x = df['bmi'].values.reshape(-1,1)
y = df['target'].values.reshape(-1,1)
```

#### Instrucción 4: 
<p style='text-align: justify;'>Divide el conjunto de datos en entrenamiento y prueba. Asigna el 80 porciento de los datos para el entrenamiento y el resto para la prueba.</p>

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

#### Instrucción 5: 
<p style='text-align: justify;'>Con ayuda de la paquetería de sklearn construye un modelo de <b>regresión polinomial de grado 2</b> que se ajuste a nuestros datos de entrenamiento y utiliza el modelo para obtener los dos siguientes conjuntos de respuestas: "<b>y_train_predict</b>" y "<b>y_test_predict</b>".</p>

<p style='text-align: justify;'>Posterior a construir el modelo utilizando la función <b>plt.scatter</b> construye una grafica donde se muestren el conjuto de entrenamiento en color azul, el conjunto de prueba en naranja y una línea recta roja donde se muestre la predicción devuelta por nuestro modelo.</p>

Por último obten la metrica r2_score para el modelo obtenido

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
poly = PolynomialFeatures(degree = 2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.fit_transform(x_test)
reg = LinearRegression().fit(x_train_poly, y_train)
y_train_predict = reg.predict(x_train_poly)
y_test_predict = reg.predict(x_test_poly)
```

```python
plt.scatter(x_train, y_train, alpha = 0.2, c = 'blue')
plt.scatter(x_test, y_test, alpha = 0.8, c = 'orange')
    
x_model = np.linspace(-0.1,0.18).reshape(-1, 1)
y_model = reg.predict(poly.transform(x_model.reshape(-1,1)))
plt.plot(x_model, y_model, "r", lw = 4)
    
plt.axis([-0.1, 0.2, 0, 360])
plt.xlabel("bmi")
plt.ylabel("target")
    
plt.show()
```

```python
print("Entrenamiento", r2_score(y_train, y_train_predict))
print("Prueba", r2_score(y_test, y_test_predict))
```
